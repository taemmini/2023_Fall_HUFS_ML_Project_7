```python
import os
import sys
import urllib.request
import requests

# 네이버 개발자 센터에서 발급받은 Client ID와 Client Secret 넣어주기
client_id = "Client ID"
client_secret = "Client Secret"

# 네이버 로그인 ID와 Password 넣어주기
naver_id = "NaverID"
naver_password = "NaverPassword"

# 네이버 로그인 API URL
naver_login_url = "https://nid.naver.com/oauth2.0/token"

# 네이버 로그인 API에 전달할 파라미터
params = {
    "grant_type": "password",
    "client_id": client_id,
    "client_secret": client_secret,
    "username": naver_id,
    "password": naver_password,
}

# 네이버 로그인 API 호출
response = requests.post(naver_login_url, data=params)
result = response.json()

# 얻은 액세스 토큰
access_token = result.get("access_token")

# 액세스 토큰을 사용하여 네이버 API 호출
url = "https://openapi.naver.com/v1/captcha/nkey?code=0"
headers = {
    "Authorization": f"Bearer {access_token}",
    "X-Naver-Client-Id": client_id,
    "X-Naver-Client-Secret": client_secret,
}

response = requests.get(url, headers=headers)
if response.status_code == 200:
    response_body = response.json()
    print(response_body)
else:
    print("Error Code:", response.status_code)
    print(response.text)
```

```python
import os
import sys
import urllib.request

client_id = "R6iPPDs3U63M64Y8DWEw"
client_secret = "6jknzgxbYd"

# 저장 폴더 경로 지정
save_folder = "C:/Users/User/Desktop/Captcha"  # 변경하고자 하는 폴더 경로로 수정

# 1000번 반복
for i in range(1000):
    key = "CaptchaKey"  # 캡차 Key 값 넣어주기
    url = "https://openapi.naver.com/v1/captcha/ncaptcha.bin?key=" + key
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id", client_id)
    request.add_header("X-Naver-Client-Secret", client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()

    if rescode == 200:
        print(f"캡차 이미지 {i + 1}번째 다운로드")
        response_body = response.read()
        with open(os.path.join(save_folder, f'captcha_{i + 1}.jpg'), 'wb') as f:
            f.write(response_body)
    else:
        print(f"Error Code({i + 1}): {rescode}")
```
